```python
import codecs
import pandas as pd
from datetime import datetime
from statsbombpy import sb
```

```python
#Load in Statsbomb competition and match data
#This is a library for loading json files.
import json

#Load the competition file
competitions = pd.read_json(codecs.open('../Statsbomb/data/competitions.json','r','utf-8'))
```

```python
competitions.head()
```

```python
laLiga = competitions[competitions['competition_name']=='La Liga'].reset_index()
```

```python
laLiga.head()
```

```python
comp_id = laLiga['competition_id'][0]
```

```python
messiYears = list(range(2004,2022))
```

```python
messiLaLiga =[]
messiSeasons =[]
for y in messiYears:
    for index, row in laLiga.iterrows():
        if(str(y) in row['season_name']):
            if(row['season_id'] in messiLaLiga):
                pass
            else:
                messiLaLiga.append(row['season_id'])
                messiSeasons.append(y)
                
            
```

```python
messiLaLiga
```

```python
messiSeasons[0]
```

```python
temp = pd.read_json(codecs.open('../Statsbomb/data/matches/11/1.json',"r","utf-8"))
```

```python
#Load the list of matches for this competition
col = temp.columns.tolist()
matches = pd.DataFrame(columns=col)
count = 0
for i in messiLaLiga:
    match = pd.read_json(codecs.open('../Statsbomb/data/matches/'+str(comp_id)+"/"+str(i)+".json","r","utf-8"))
    match['season'] = messiSeasons[count]
    count = count+1
    matches = pd.concat([matches, match])
```

```python
matches.head(1)
```

```python
matches = matches.drop(['kick_off','competition','match_status','match_status_360','last_updated','last_updated_360','metadata','competition_stage'],axis=1)
```

```python
matches['match_date'] = pd.to_datetime(matches['match_date'],format="%Y-%m-%d")
```

```python
matches['home_team'].iloc[0]
        
```

```python
manager=[]
oppManager=[]
homeTeam = []
awayTeam = []
stadium =[]
referee=[]
for index,row in matches.iterrows():
    try:
        if(row['home_team']['home_team_id']==217):
            manager.append(row['home_team']['managers'][0]['name'])
        else:
            manager.append(row['away_team']['managers'][0]['name'])
    except:
        manager.append('NA')
    try:
        if(row['home_team']['home_team_id']==217):
            oppManager.append(row['away_team']['managers'][0]['name'])
        else:
            oppManager.append(row['home_team']['managers'][0]['name'])            
    except:
        oppManager.append('NA')
    
    homeTeam.append(row['home_team']['home_team_name'])
    awayTeam.append(row['away_team']['away_team_name'])
    stadium.append(row['stadium']['name'])
    try:
        referee.append(row['referee']['name'])
    except:
        referee.append('NA')
    
matches['home_team_name'] = homeTeam
matches['away_team_name'] = awayTeam
matches['manager'] = manager
matches['opp_manager'] = oppManager
matches['stadium'] = stadium
matches['referee'] = referee
```

```python
matches[matches['manager']=='NA'].head()
```

```python
count=0
for index,row in matches.iterrows():
    if(row['home_team_name']=='Barcelona'):
        if(row['manager']=='NA'):
            if(manager[count-1] != 'NA' and manager[count+1]!='NA' and manager[count-1]==manager[count+1]):
                manager[count] = manager[count-1]
    else:
        if(row['manager']=='NA'):
            if(manager[count-1] != 'NA' and manager[count+1]!='NA' and manager[count-1]==manager[count+1]):
                manager[count] = manager[count-1]
        
    count=count+1
```
